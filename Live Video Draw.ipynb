```python
import cv2
```

```python
pt1, pt2 = (0, 0), (0, 0)
top_left, bottom_right = False, False

def set_hyper(event, x, y, flags, param):
    global pt1, pt2, top_left, bottom_right

    if event == cv2.EVENT_LBUTTONDOWN:
        if not top_left:
            pt1 = (x, y)
            top_left = True
            print("Top-left:", pt1)

        elif not bottom_right:
            pt2 = (x, y)
            bottom_right = True
            print("Bottom-right:", pt2)

        else:
            # Reset on 3rd click
            pt1, pt2 = (0, 0), (0, 0)
            top_left, bottom_right = False, False
            print("Reset points")
```

```python
cap = cv2.VideoCapture(0)

cv2.namedWindow("Stream")
cv2.setMouseCallback("Stream", set_hyper)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Draw top-left point
    if top_left:
        cv2.circle(
            frame,
            center=pt1,
            radius=5,
            color=(0, 0, 255),
            thickness=-1
        )

    # Draw rectangle when both points exist
    if top_left and bottom_right:
        cv2.rectangle(
            frame,
            pt1,
            pt2,
            color=(0, 255, 0),
            thickness=2
        )

    cv2.imshow("Stream", frame)

    # Quit on 'q' or window close
    if ((cv2.waitKey(1) & 0xFF) == ord("q")) or (cv2.getWindowProperty("Stream", cv2.WND_PROP_VISIBLE) < 1):
        break

cap.release()
cv2.destroyAllWindows()
```
