```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
import cv2
```

```python
int_ext = cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\internal_external.png", flags= cv2.IMREAD_GRAYSCALE)

plt.imshow(int_ext, cmap= "gray")
plt.axis("off")
plt.show()
```

```python
result = cv2.findContours(int_ext, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
contours = result[-2]
hierarchy = result[-1]
```

```python
display(type(contours),type(hierarchy))
```

```python
display(contours, hierarchy)
```

```python
external_contour = np.zeros(int_ext.shape)
external_contour.shape
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contour, contours, i,255,-1)
plt.imshow(external_contour, cmap="gray")
```

```python
internal_contour = np.zeros(int_ext.shape)
internal_contour.shape
for i in range(len(contours)):
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contour, contours, i,255,-1)
plt.imshow(internal_contour, cmap="gray")
```
