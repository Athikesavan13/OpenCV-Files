```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def load_img():
  img = cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/bricks.jpg").astype(np.float32) / 255
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  return img
```

```python
img1 = load_img()
```

```python
def display_img(img):
  fig = plt.figure(figsize=(8,8))
  ax = fig.add_subplot(111)
  ax.imshow(img)
```

```python
display_img(img1)
```

```python
gamma = 1/ 5
new_img1 = np.power(img1,gamma)
display_img(new_img1)
```

```python
gamma = 2
new_img1 = np.power(img1,gamma)
display_img(new_img1)
```

```python
img2 = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img2,text = "Bricks", org=(50,600),fontFace=font, fontScale=10,color=(255,0,0), thickness= 10)
display_img(img2)
```

```python
kernel = np.ones((5,5),dtype= np.float32)/25
dst = cv2.filter2D(img2, ddepth=-1,kernel=kernel)
display_img(dst)
```
