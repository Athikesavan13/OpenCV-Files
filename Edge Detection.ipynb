```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\sammy_face.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(img, 127,127)
plt.imshow(edges)
```

```python
med_val1 = np.median(edges)
med_val2 = np.median(img)
display(med_val1, med_val2)
```

```python
lower = int(max(0,0.7*med_val2))
upper = int(min(255,1.3*med_val2))
```

```python
display(lower, upper)
```

```python
edges = cv2.Canny(img, lower, upper+150)
plt.imshow(edges)
```

```python
blurred = cv2.GaussianBlur(img, (7,7), sigmaX=0.5, sigmaY=0.8)
plt.imshow(blurred)
```

```python
edges = cv2.Canny(blurred, lower+75, upper+175)
plt.imshow(edges)
```
