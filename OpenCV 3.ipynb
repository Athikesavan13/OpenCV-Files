```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
black_img = np.zeros(shape = (512,512,3), dtype = np.int16)
```

```python
black_img.shape
```

```python
plt.figure(figsize=(3,3))
plt.imshow(black_img)
```

```python
rect_img = cv2.rectangle(black_img, pt1 = (300,25), pt2 = (450,150), color = (255,0,10), thickness = 10)
```

```python
plt.figure(figsize=(3,3))
plt.imshow(rect_img)
```

```python
rect1_img = cv2.rectangle(black_img, pt1 = (200,200), pt2 = (250,250), color = (100,255,100), thickness = 10)
plt.figure(figsize=(3,3))
plt.imshow(rect1_img)
```

```python
circle_img = cv2.circle(black_img,center = (100,100),radius = 30,color = (255,255,255), thickness = 10)
plt.figure(figsize=(3,3))
plt.imshow(circle_img)
```

```python
circle_img = cv2.circle(black_img,center = (100,100),radius = 30,color = (255,255,255), thickness = 10)
plt.figure(figsize=(3,3))
plt.imshow(circle_img)
```

```python
polygon_img = cv2.line(black_img,pt1=(0,0),pt2=(200,200),color = (255,255,255), thickness = 10)
plt.figure(figsize=(3,3))
plt.imshow(polygon_img)
```

```python
blank1 = np.zeros(shape = (512,512,3), dtype = np.int16)
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank1, text = "hello", org = (100,100),color = (255,255,255), fontFace= font, fontScale= 4, thickness= 2)
plt.imshow(blank1)
```

polygon

```python
vertices = np.array([ [0,0], [100,100], [200,200],[300,300],[200,400],[100,500], [0,0]], dtype= np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
pts
```

```python
blank2 = np.zeros(shape = (512,512,3), dtype = np.int16)
cv2.polylines(blank2,[pts],isClosed= True,color= (255,0,0), thickness= 4)
plt.imshow(blank2)
```
