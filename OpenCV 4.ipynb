```python
img1 = cv2.cvtColor(cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/dog_backpack.png"),cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/watermark_no_copy.png"),cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(5,3))

ax[0].imshow(img1)
ax[0].set_title("Image 1")
ax[0].axis("off")

ax[1].imshow(img2)
ax[1].set_title("Image 2")
ax[1].axis("off")

plt.show()
```

Blending Images of same size

```python
s_img1 = cv2.resize(img1,(1200,1200))
s_img2 = cv2.resize(img2,(1200,1200))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(5,3))

ax[0].imshow(s_img1)
ax[0].set_title("Image 1")
ax[0].axis("off")

ax[1].imshow(s_img2)
ax[1].set_title("Image 2")
ax[1].axis("off")

plt.show()
```

```python
blended = cv2.addWeighted(src1 = s_img1, alpha= 1.0, src2 = s_img2, beta= 0.4, gamma= 0)
plt.figure(figsize=(3,3))
plt.imshow(blended)
```

Different Image Sizes and pixel density


```python
d_img1= img1.copy()
d_img2 = cv2.resize(img2, (600,600))
```

```python
d_img1.shape
```

```python
d_img2.shape
```

```python
large_img , small_img = d_img1.copy(),d_img2.copy()
```

```python
x_offset, y_offset = 0,0
x_end, y_end = small_img.shape[0],small_img.shape[1]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```

```python
blended = cv2.addWeighted(src2 = large_img, alpha= 0.2, src1 = d_img1, beta= 1, gamma= 0)
plt.figure(figsize=(3,3))
plt.imshow(blended)
```
