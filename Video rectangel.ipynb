```python
import cv2
import numpy
```

```python


cap = cv2.VideoCapture(0)
cv2.namedWindow("Stream")

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

l_s, b_s = height // 2, width // 2
l_e, b_e = l_s + (height // 4), b_s + (width // 4)

while cap.isOpened():
    rt, frame1 = cap.read()
    if not rt:
        break   

    cv2.rectangle(frame1, (b_s, l_s), (b_e, l_e), (0, 0, 255), 2)
    cv2.imshow("Stream", frame1)

    if ((cv2.waitKey(1) & 0xFF) == ord("q"))  or (cv2.getWindowProperty("Stream", cv2.WND_PROP_VISIBLE) < 1):
        break

cap.release()
cv2.destroyAllWindows()
```
