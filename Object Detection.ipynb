```python
flat_chess = cv2.cvtColor(cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\flat_chessboard.png"), cv2.COLOR_BGR2RGB)
real_chess = cv2.cvtColor(cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\real_chessboard.jpg"), cv2.COLOR_BGR2RGB)
gray_flat = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
gray_real = cv2.cvtColor(real_chess, cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat, maxCorners= 64, qualityLevel= 0.01, minDistance= 10)
corners = np.int32(corners)
display(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess, (x,y), 3, (255,0,0), -1)
```
