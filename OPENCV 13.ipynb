```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2

pennies = cv2.cvtColor(
    cv2.imread('/content/drive/MyDrive/Computer-Vision-with-Python/DATA/pennies.jpg'),
    cv2.COLOR_BGR2RGB
)
```

```python
def display(image, cmap):
  if cmap is not None :
    plt.figure(figsize = (10,10))
    plt.imshow(image, cmap = 'gray')
  else:
    plt.figure(figsize = (10,10))
    plt.imshow(image)
```

```python
display(pennies, None)
```

```python
sep_blur = cv2.medianBlur(pennies, ksize= 21)
```

```python
sep_blur_gray_img = cv2.cvtColor(sep_blur, cv2.COLOR_RGB2GRAY)
```

```python
ret, thresh_sep = cv2.threshold(sep_blur_gray_img, 170,255, cv2.THRESH_BINARY_INV)
display(thresh_sep, 'gray')
```

```python
contours, hierarchy = cv2.findContours(
    thresh_sep.copy(),
    cv2.RETR_CCOMP,
    cv2.CHAIN_APPROX_SIMPLE
)
```

```python
pennies_copy = pennies.copy()
```

```python
for i in range (len(contours)):
  if hierarchy[0][i][3] == -1 :
    cv2.drawContours(pennies_copy, contours, i, (255,0,0), 15)
```
