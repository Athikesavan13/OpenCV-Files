```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
import cv2
```

```python
chess_board = cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\flat_chessboard.png")

plt.imshow(chess_board)
plt.axis("off")
plt.show()
```

```python

dot_grid = cv2.imread("D:\\Python files\\Computer-Vision-with-Python\\DATA\\dot_grid.png")
dot_grid = cv2.cvtColor(dot_grid, cv2.COLOR_BGR2RGB)

plt.imshow(dot_grid)
plt.axis("off")
plt.show()
```

```python
found , corners = cv2.findChessboardCorners(chess_board, (7,7))
display(found)
if found:
    display(corners, corners.shape)
```

```python
vis = chess_board.copy()

cv2.drawChessboardCorners(vis, (7, 7), corners, found)

plt.imshow(cv2.cvtColor(vis, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

```python
pts = corners.squeeze()  # shape  (49, 2)

plt.scatter(pts[:,0], pts[:,1], c='r')
plt.gca().invert_yaxis()
plt.title("Detected inner corners")
plt.show()
```

```python
found, centre = cv2.findCirclesGrid(dot_grid,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
cv2.drawChessboardCorners(dot_grid,(10,10), centre, found)
plt.imshow(dot_grid)
```
