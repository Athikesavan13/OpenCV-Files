```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
cap = cv2.VideoCapture("D:\Python files\OpenCV\Trial1.mp4")
cv2.namedWindow("Video Viewer")
if cap.isOpened() == False:
    print("Not A Valid File Check Codec or file location")
while cap.isOpened():
    rt,frame1 = cap.read()
    if not rt:
        break
    time.sleep(1/20)
    cv2.imshow("Video Viewer",frame1)

    if (cv2.waitKey(1) & 0xFF) == ord("q")  or cv2.getWindowProperty("Video Viewer", cv2.WND_PROP_VISIBLE) < 1:
        break

cap.release()
cv2.destroyAllWindows()
```
