```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img):
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot(1,1,1)
  ax.imshow(img)
```

```python
dark_horse = cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/horse.jpg")
show_horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)
```

```python
rainbow = cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
```

```python
blue_bricks = cv2.imread("/content/drive/MyDrive/Computer-Vision-with-Python/DATA/bricks.jpg")
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
display_img(blue_bricks)
```

```python
display_img(dark_horse)
```

```python
display_img(rainbow)
```

```python
hist_values = cv2.calcHist([blue_bricks], channels= [0], mask = None, histSize=[256], ranges= [0,256])
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([dark_horse], channels= [2], mask = None, histSize=[256], ranges= [0,256])
plt.plot(hist_values)
```

```python
img = blue_bricks
color = ('Blue', 'Green', 'Red')
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)
ax.set_xlim([0,256])
for i,colour in enumerate(color):
  hist_values = cv2.calcHist([blue_bricks], channels= [i], mask = None, histSize=[256], ranges= [0,256])
  ax.plot(hist_values,color = colour)
plt.title("Histogram For Blue Bricks")
```

Histogram Equaliztion

```python
img = rainbow
img.shape
```

```python
mask = np.zeros(img.shape[:2],dtype=np.uint8)
```

```python
print(type(mask))
print(mask.dtype)
print(mask.shape)
```

```python

plt.imshow(mask)
```

```python
mask[300:400, 100:400] = 255
```

```python
plt.imshow(mask, cmap = "gray")
```

```python
plt.imshow(show_rainbow)
```

```python
masked_img = cv2.bitwise_and(img,img, mask = mask)
```

```python
plt.imshow(masked_img)
```

```python
show_rainbow_masked_img =cv2.bitwise_and(show_rainbow, show_rainbow,
                                         mask = mask)
```

```python
plt.imshow(show_rainbow_masked_img)
```

```python
hist_mask_values = cv2.calcHist([rainbow], channels=[2], mask=mask, histSize=[256], ranges=[0,256])
hist_no_mask_values =  cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0,256])
```

```python
fig = plt.figure(figsize=(25,5))

ax = fig.add_subplot(1,1,1)
ax.plot(hist_mask_values, color = "red")
ax.plot(hist_no_mask_values, color = "green")
ax.set_xlim([0,256])
ax.set_ylim([0,2000])
plt.show()
```
